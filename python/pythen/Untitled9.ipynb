```python

umbrella = input("Do you have an umbrella? Enter y/n : ").lower() == "y"
raining = input("Is it raining? Enter y/n: ").lower() == "y"


if umbrella or not raining:
    print("You can go outside.")
else:
    print("Stay indoors!")
```

```python

is_sunny = input("Is it sunny? Enter y/n : ").lower() == "y"
is_cloudy = input("Is it cloudy? Enter y/n : ").lower() == "y"


if (is_sunny and not is_cloudy) or (not is_sunny and is_cloudy):
    print("Weather is uncertain.")
else:
    print("Weather is clear.")
```

```python

is_sunny = input("Is it sunny? Enter y/n : ").lower() == "y"
has_umbrella = input("Do you have an umbrella? Enter y/n : ").lower() == "y"


if is_sunny and has_umbrella:
    print("You can enjoy a sunny day with your umbrella!")
else:
    print("Consider taking your umbrella just in case.")
```
