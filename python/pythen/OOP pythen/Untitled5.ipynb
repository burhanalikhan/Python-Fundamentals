```python
import tkinter as tk
from tkinter import messagebox
class Calculator:
    def __init__(self):
        self.num=0
        self.mnum=1
        self.act=99
        self.window=tk.Tk()
        self.window.geometry("500x450")
        self.window.title("Pyhten GUI")
        
        self.label=tk.Label(self.window,text="Hello Burhan",font=('Arial',20))
        self.label.pack(padx=20,pady=20)
        
        self.textbox=tk.Text(self.window,height=1,font=('Arial',20))
        self.textbox.pack(padx=10,pady=10)
        
        self.passw=tk.Entry(self.window,font=('Arial',20))
        self.passw.insert(tk.END,self.num)
        self.passw.pack(padx=40,pady=10)
        
        self.buttonframe=tk.Frame(self.window)
        self.buttonframe.columnconfigure(0,weight=1)
        self.buttonframe.columnconfigure(1,weight=1)
        self.buttonframe.columnconfigure(2,weight=1)
        self.buttonframe.columnconfigure(3,weight=1)
        self.buttonframe.columnconfigure(4,weight=1)
        
        
        self.btn1=tk.Button(self.buttonframe,text=1,font=('Arial',20),command=self.one)
        self.btn1.grid(row=0,column=0,sticky=tk.W+tk.E)
        
        self.btn2=tk.Button(self.buttonframe,text=2,font=('Arial',20),command=self.two)
        self.btn2.grid(row=0,column=1,sticky=tk.W+tk.E)
        
        self.btn3=tk.Button(self.buttonframe,text=3,font=('Arial',20),command=self.three)
        self.btn3.grid(row=0,column=2,sticky=tk.W+tk.E)
        
        self.btnadd=tk.Button(self.buttonframe,text='+',font=('Arial',20),command=self.add)
        self.btnadd.grid(row=0,column=3,sticky=tk.W+tk.E)
        
        self.btnminus=tk.Button(self.buttonframe,text='-',font=('Arial',20),command=self.minus)
        self.btnminus.grid(row=0,column=4,sticky=tk.W+tk.E)
        
        self.btn4=tk.Button(self.buttonframe,text=4,font=('Arial',20),command=self.four)
        self.btn4.grid(row=1,column=0,sticky=tk.W+tk.E)
        
        self.btn5=tk.Button(self.buttonframe,text=5,font=('Arial',20),command=self.five)
        self.btn5.grid(row=1,column=1,sticky=tk.W+tk.E)
        
        self.btn6=tk.Button(self.buttonframe,text=6,font=('Arial',20),command=self.six)
        self.btn6.grid(row=1,column=2,sticky=tk.W+tk.E)
        
        self.btndiv=tk.Button(self.buttonframe,text='/',font=('Arial',20),command=self.div)
        self.btndiv.grid(row=1,column=3,sticky=tk.W+tk.E)
        
        self.btnmul=tk.Button(self.buttonframe,text='x',font=('Arial',20),command=self.mult)
        self.btnmul.grid(row=1,column=4,sticky=tk.W+tk.E)
        
        self.btn7=tk.Button(self.buttonframe,text=7,font=('Arial',20),command=self.seven)
        self.btn7.grid(row=2,column=0,sticky=tk.W+tk.E)
        
        self.btn8=tk.Button(self.buttonframe,text=8,font=('Arial',20),command=self.eight)
        self.btn8.grid(row=2,column=1,sticky=tk.W+tk.E)
        
        self.btn9=tk.Button(self.buttonframe,text=9,font=('Arial',20),command=self.nine)
        self.btn9.grid(row=2,column=2,sticky=tk.W+tk.E)
        
        self.btnans=tk.Button(self.buttonframe,text='=',font=('Arial',20),command=self.ans)
        self.btnans.grid(row=2,column=3,sticky=tk.W+tk.E)
        
        self.btncl=tk.Button(self.buttonframe,text='C',font=('Arial',20),command=self.clear)
        self.btncl.grid(row=2,column=4,sticky=tk.W+tk.E)
        
        self.btn0=tk.Button(self.buttonframe,text=0,font=('Arial',20),command=self.zero)
        self.btn0.grid(row=3,column=1,sticky=tk.W+tk.E)
        
        self.buttonframe.pack(fill='x')
        
        
        self.window.mainloop()
    def one(self):
        self.textbox.insert(tk.END,'1')
    def two(self):
        self.textbox.insert(tk.END,'2')
    def three(self):
        self.textbox.insert(tk.END,'3')
    def four(self):
        self.textbox.insert(tk.END,'4')
    def five(self):
        self.textbox.insert(tk.END,'5')
    def six(self):
        self.textbox.insert(tk.END,'6')
    def seven(self):
        self.textbox.insert(tk.END,'7')
    def eight(self):
        self.textbox.insert(tk.END,'8')
    def nine(self):
        self.textbox.insert(tk.END,'9')
    def zero(self):
        self.textbox.insert(tk.END,'0')
    def clear(self):
        self.num=0
        self.act=99
        self.mnum=1
        self.textbox.delete('1.0', tk.END)
        self.passw.delete(0, tk.END)
    def add(self):
        self.act=6
        self.num=float(self.textbox.get('1.0', tk.END))+self.num
        self.textbox.delete('1.0', tk.END)
        self.passw.delete(0, tk.END)
        self.passw.insert(tk.END,self.num)
    def minus(self):
        if self.act !=7:
            self.act=7
            self.num=float(self.textbox.get('1.0', tk.END))-self.num
            self.textbox.delete('1.0', tk.END)
            self.passw.delete(0, tk.END)
            self.passw.insert(tk.END,self.num)
        elif self.act==7:
            self.num=self.num-float(self.textbox.get('1.0', tk.END))
            self.textbox.delete('1.0', tk.END)
            self.passw.delete(0, tk.END)
            self.passw.insert(tk.END,self.num)
    def mult(self):
        self.act=8
        self.mnum=float(self.textbox.get('1.0', tk.END))*self.mnum
        self.textbox.delete('1.0', tk.END)
        self.passw.delete(0, tk.END)
        self.passw.insert(tk.END,self.mnum)
    def div(self):
        if self.act != 9:
            self.act=9
            self.mnum=float(self.textbox.get('1.0', tk.END))/self.mnum
            self.textbox.delete('1.0', tk.END)
            self.passw.delete(0, tk.END)
            self.passw.insert(tk.END,self.mnum)
        elif self.act==9:
            self.mnum=self.mnum/float(self.textbox.get('1.0', tk.END))
            self.textbox.delete('1.0', tk.END)
            self.passw.delete(0, tk.END)
            self.passw.insert(tk.END,self.mnum)
    def ans(self):
        if self.act == 6:
            self.num=float(self.textbox.get('1.0', tk.END))+self.num
            self.textbox.delete('1.0', tk.END)
            self.textbox.insert(tk.END,self.num)
            self.num=0
            self.act=99
            self.passw.delete(0, tk.END)
        elif self.act==7:
            self.num=self.num-float(self.textbox.get('1.0', tk.END))
            self.textbox.delete('1.0', tk.END)
            self.textbox.insert(tk.END,self.num)
            self.num=0
            self.act=99
            self.passw.delete(0, tk.END)
        elif self.act==8:
            self.mnum=float(self.textbox.get('1.0', tk.END))*self.mnum
            self.textbox.delete('1.0', tk.END)
            self.textbox.insert(tk.END,self.mnum)
            self.mnum=1
            self.act=99
            self.passw.delete(0, tk.END)
        elif self.act==9:
            self.mnum=self.mnum/float(self.textbox.get('1.0', tk.END))
            self.textbox.delete('1.0', tk.END)
            self.textbox.insert(tk.END,self.mnum)
            self.mnum=1
            self.act=99
            self.passw.delete(0, tk.END)
        elif self.act==99:
            messagebox.showinfo(title="Message",message="You didnt select any action")
cal=Calculator()
```
