```python
from abc import ABC, abstractmethod

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    
    def total_cost(self):
        return self.__price*self.__quantity

class Beverage(Product):
    def __init__(self, name, price, quantity, volume):
        super().__init__(name, price, quantity)
        self.volume = volume
    
    def total_cost(self):
        return self.price * self.volume

class Snack(Product):
    def __init__(self, name, price, quantity, weight):
        super().__init__(name, price, quantity)
        self.weight = weight
    
    def total_cost(self):
        return self.price * self.weight

class DairyProduct(Product):
    def __init__(self, name, price, quantity, expiry_date):
        super().__init__(name, price, quantity)
        self.expiry_date = expiry_date
    
    def total_cost(self):
        return self.price * self.quantity

class Confectionery(Product):
    def __init__(self, name, price, quantity, sugar_content):
        super().__init__(name, price, quantity)
        self.sugar_content = sugar_content
    
    def total_cost(self):
        return self.price * self.quantity

class InventoryOperations(ABC):
    @abstractmethod
    def add_product(self, product):
        pass
    
    @abstractmethod
    def update_inventory(self, product_name, new_quantity):
        pass
    
    @abstractmethod
    def generate_report(self):
        pass

class InventoryManager(InventoryOperations):
    def __init__(self):
        self.inventory = {}
    
    def add_product(self, product):
        if product.name in self.inventory:
            self.inventory[product.name].quantity += product.quantity
            print("Quantity updated")
        else:
            self.inventory[product.name] = product
            print("New Product added")
        
    
    def update_inventory(self, product_name):
        if product_name in self.inventory:
            update=input("What you want to update p for Price or q for Quantity and b for both")
            if update=="p":
                Price=float(input("Please enter new price: "))
                self.inventory[product_name].price = Price
                print("Product updated")
            elif update=="q":
                Quantity=float(input("Please enter new quantity: "))
                self.inventory[product_name].quantity = Quantity 
                print("Product updated")
            elif update=="b" :
                Price=float(input("Please enter new price: "))
                Quantity=float(input("Please enter new quantity: "))
                self.inventory[product_name].price = Price
                self.inventory[product_name].quantity = Quantity
                print("Product updated")
        else:
            print("Product not found in inventory.")
    
    def generate_report(self):
        for product_name, product in self.inventory.items():
            print(f"Product: {product.name}, Price: ${product.price}, Quantity: {product.quantity}")

# Example Usage
beverage1 = Beverage("Coca-Cola", 1.5, 100, 330)
snack1 = Snack("Potato Chips", 2.0, 50, 100)
dairy_product1 = DairyProduct("Milk", 3.0, 20, "2024-04-01")
confectionery1 = Confectionery("Chocolate Bar", 1.0, 200, "High")

im = InventoryManager()
im.add_product(beverage1)
im.add_product(snack1)
im.add_product(dairy_product1)
im.add_product(confectionery1)


```
