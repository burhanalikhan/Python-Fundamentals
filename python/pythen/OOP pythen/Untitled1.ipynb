```python
class Animal:
    def __init__(self,name="",species=""):
        self.__name=name
        self.__species=species
    def eat(self):
        print(f"{self._Animal__name} is eating")
        
class Pet(Animal):
    def __init__(self,name="",species="",breed=""):
        super().__init__(name,species)
        self.__breed=breed
    def play(self):
        print(f"{self._Animal__name} is playing")
        
class Cat(Pet):
    def __init__(self,name="",species="",breed="",ids=""):
        super().__init__(name,species,breed)
        self.__ids=ids
    def meow(self):
        print("Meow Meow")
c=Cat("Tom","german","japan","8")
c.meow()
c.play()
c.eat()
```

```python

# Function to take multiple arguments
def add(datatype, *y):
 
    # if datatype is int
    # initialize answer as 0
    if datatype == 'int':
        answer = 0
 
    # if datatype is str
    # initialize answer as ''
    if datatype == 'str':
        answer = ''
 
 # Traverse through the arguments
    for x in y:
 
        # This will do addition if the
        # arguments are int. Or concatenation
        # if the arguments are str
        answer = answer + x
 
    print(answer)
 
 
# Integer
add('int', 5, 6)
 
# String
add('str', 'Hi ', 'Geeks')
```
