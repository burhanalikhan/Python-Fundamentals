```python
class Item:
    pay_rate=0.8
    all =[]
    def __init__(self,name : str,price : float,quantity=0):
        assert price>=0
        assert quantity>=0
        
        self.__name=name
        self.__price=price
        self.__quantity=quantity
        
        Item.all.append(self)
        
    @property
    def name(self):
        return self.__name
    @property
    def price(self):
        return self.__price
    @property
    def quantity(self):
        return self.__quantity
    
    @name.setter
    def name(self,value):
        self.__name=value
        
    def total_price(self):
        return self.__price*self.__quantity


item1 = Item("market", 700, 2)

# Attempting to modify the name using the setter
item1.name = "Mark"

# Printing the name
print(item1.total_price())
```

```python
class Phone(Item):
    def __init__(self,name:str,price:float,quantity=0,brokenphone=0):
        super().__init__(name,price,quantity)
        assert brokenphone>=0
        self.__brokenphone=brokenphone
    @property
    def brokenphone(self):
        return self.__brokenphone
```

```python
class Person:
    def __init__(self,name : str,phone_no :int,dob :int):
        self.name=name
        self.phone_no=phone_no
        self.dob=dob
class Professor(Person):
    def __init__(self,name:str,phone_no: int,dob: int,room: int):
        super().__init__(name,phone_no,dob)
        self.room=room
    def display_room(self):
        print(f"{self.name} Room: {self.room}")
        
pr1=Professor("Burhan",331,3,5)
pr1.display_room()
```

```python
class Laptop(Phone):
    def __init__(self, name: str, price: float, quantity=0, brokenphone=0, quality=""):
        super().__init__(name, price, quantity, brokenphone)
        self.__quality = quality

    @property
    def quality(self):
        return self.__quality
    
    def display_info(self):
        print(f"Name: {self.name} Price: {self.price} Quantity: {self.quantity} Quality: {self.quality}")


lap = Laptop("Lenovo", 3000, 3, 1, "High")
lap.display_info()
```
