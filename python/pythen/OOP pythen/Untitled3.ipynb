```python
class Person:
    def __init__(self,name,date_ofbirth):
        self.__name=name
        self.__date_ofbirth=date_ofbirth
    def display(self):
        print(f"Name : {self.__name}   \nDate of Birth : {self.__date_ofbirth} " )
    

class Teacher(Person):
    def __init__(self,name,date_ofbirth,ids,designation):
        super().__init__(name,date_ofbirth)
        self.__ids=ids
        self.__designation=designation
    def display(self):
        print(f"Name : {self._Person__name} \nDate of Birth : {self._Person__date_ofbirth} \nId : {self.__ids}  \nDesignation : {self.__designation}")
        
class Student(Person):
    def __init__(self,name,date_ofbirth,clas,ids):
        super().__init__(name,date_ofbirth)
        self.__clas=clas
        self.ids=ids
    def display(self):
        print(f"Name : {self._Person__name} \nDate of Birth : {self._Person__date_ofbirth} \nId : {self.ids}  \nClass : {self.__clas}")

s2=Student("Burhan","20-5-2005","BSE-2B","02-131232-073")
s2.display()

```

```python
class Number:
    def __init__(self, number):
        self.number = number
        
    def even(self, num1):
        for c in range(self.number, num1):
            print(f"Even numbers are: {c*2}")
a=Number(4)
a.even(14)
```
