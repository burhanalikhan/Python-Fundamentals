```python
from abc import ABC ,abstractmethod
class Shape(ABC):
    @abstractmethod
    def calculateArea(self):
        pass  # Abstract method to be overridden by subclasses

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def calculateArea(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def calculateArea(self):
        return self.width * self.height

class ShapeManager:
    def __init__(self):
        self.shapes = []

    def addShape(self, shape):
        self.shapes.append(shape)

    def removeShape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def displayShapes(self):
        for shape in self.shapes:
            print("Area of shape:", shape.calculateArea())

# Usage
if __name__ == "__main__":
    # Create some shapes
    circle1 = Circle(5)
    rectangle1 = Rectangle(4, 6)
    circle2 = Circle(3)
    

    # Create a ShapeManager
    manager = ShapeManager()

    # Add shapes to the manager
    manager.addShape(circle1)
    manager.addShape(rectangle1)
    manager.addShape(circle2)

    # Display areas of all shapes
    manager.displayShapes()
    manager.removeShape(circle1)
    circle2.radius=4

    # Display areas of remaining shapes
    manager.displayShapes()
```

```python
# Interface for shapes
class Shape:
    def calculateArea(self):
        pass

    def displayShapeInfo(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculateArea(self):
        return 3.14 * self.radius * self.radius

    def displayShapeInfo(self):
        print("Circle - Radius:", self.radius)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculateArea(self):
        return self.width * self.height

    def displayShapeInfo(self):
        print("Rectangle - Width:", self.width, "Height:", self.height)

# ShapeCollection class to manage shapes using composition
class ShapeCollection:
    def __init__(self):
        self.shapes = []

    def addShape(self, shape):
        self.shapes.append(shape)

    def removeShape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def displayShapes(self):
        for shape in self.shapes:
            shape.displayShapeInfo()
            print("Area:", shape.calculateArea())

# Usage
if __name__ == "__main__":
    # Create some shapes
    circle1 = Circle(5)
    rectangle1 = Rectangle(4, 6)
    circle2 = Circle(3)
    
    
    # Create a ShapeCollection
    collection = ShapeCollection()

    # Add shapes to the collection
    collection.addShape(circle1)
    collection.addShape(rectangle1)
    collection.addShape(circle2)

    # Display information about all shapes
    collection.displayShapes()
    
    
    # Remove a shape
    collection.removeShape(circle1)

    # Display information about remaining shapes
    collection.displayShapes()
```
