```python
import tkinter as tk
from tkinter import messagebox

class GUI:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Python GUI")
        self.window.geometry("600x500")

        self.label = tk.Label(self.window, text="Your Message", font=('Arial', 20))
        self.label.pack(padx=10, pady=10)

        self.textbox = tk.Text(self.window, height=5, font=('Arial', 16))
        self.textbox.pack(padx=10, pady=10)

        self.check_stat = tk.IntVar()
        self.check = tk.Checkbutton(self.window, text="Show Message", font=('Arial', 16), variable=self.check_stat)
        self.check.pack(padx=10, pady=10)

        self.btn = tk.Button(self.window, text="Show Message", font=('Arial', 16), command=self.show_message, underline=0)
        self.btn.pack(padx=10, pady=10)

        # Bind Alt+S to show_message
        self.window.bind_all("<Alt-s>", self.alt_s_pressed)

        self.window.mainloop()

    def show_message(self):
        if self.check_stat.get() == 0:
            print(self.textbox.get('1.0', tk.END))
        else:
            messagebox.showinfo(title="Message", message=self.textbox.get('1.0', tk.END))

    def alt_s_pressed(self, event):
        self.show_message()

gui = GUI()
```

```python
import tkinter as tk
from tkinter import messagebox

class MnemonicApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Mnemonic Example")

        # Creating a menu
        menubar = tk.Menu(root)
        root.config(menu=menubar)

        # Adding a file menu with a mnemonic
        file_menu = tk.Menu(menubar, tearoff=0)
        menubar.add_cascade(label="File", menu=file_menu, underline=0)
        
        # Adding commands to the file menu with mnemonics
        file_menu.add_command(label="Open", underline=0, command=self.open_file, accelerator="Alt+O")
        file_menu.add_command(label="Save", underline=0, command=self.save_file, accelerator="Alt+S")
        file_menu.add_separator()
        file_menu.add_command(label="Exit", underline=1, command=root.quit, accelerator="Alt+X")
        
        # Binding the Alt key combinations to functions
        root.bind_all("<Alt-o>", lambda event: self.open_file())
        root.bind_all("<Alt-s>", lambda event: self.save_file())
        root.bind_all("<Alt-x>", lambda event: root.destroy())

        # Creating a label and a button with mnemonics
        self.label = tk.Label(root, text="Press Alt+O to Open, Alt+S to Save, Alt+X to Exit", font=('Arial', 12))
        self.label.pack(pady=20)

        self.button = tk.Button(root, text="Click Me", underline=0, command=self.on_button_click)
        self.button.pack(pady=10)

        # Binding the Alt key combination to the button click
        root.bind_all("<Alt-c>", lambda event: self.on_button_click())

    def open_file(self):
        messagebox.showinfo("Open", "Open File Dialog")

    def save_file(self):
        messagebox.showinfo("Save", "Save File Dialog")

    def on_button_click(self):
        messagebox.showinfo("Button", "Button Clicked")

# Creating the main application window
root = tk.Tk()
app = MnemonicApp(root)

# Running the application
root.mainloop()
```
