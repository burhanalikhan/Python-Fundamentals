```python
#Euclidean Algoritham
class gcd:
    a=0
    b=0
    def taking_GCD(self,num1:int,num2:int)-> int:
        
        self.value_giver(num1,num2)
            
        while self.b:
            self.a, self.b = self.b, self.a % self.b
        return self.a
        
    def value_giver(self,num1:int,num2:int):
        if num1>num2:
            self.a=num1
            self.b=num2
        else:
            self.a=num2
            self.b=num1
        
b=gcd()
c=b.taking_GCD(21,17)
print(f"{c}")

```

```python
#constructing base b expansion
class base:
    def __init__(self):
        self.q=0
        self.a=[]
    def base_converter(self,n:int,b:int):
        self.q=n
        while self.q:
            self.a.append(self.q%b)
            self.q //= b
        return self.a[::-1]

    
```

```python
#moduler expansion
class mode:
    def __init__(self):
        self.v=1
        self.power=0
    def expantion(self,b,p,m):
        x=base()
        d=x.base_converter(p,2)
        d=d[::-1]
        self.power=b % m
        for i in d:
            if i==1:
                self.v=(self.v*self.power) % m
            self.power=(self.power*self.power) % m
        return self.v
u=mode()
w=u.expantion(3,2003,99)
print(f"{w}")
```

```python
#Linear search
class linear_Search:
    def __init__(self):
        self.i=0
    def search_num(self,n,a:list)-> int:
        for self.i in range(len(a)):
            if a[self.i] == n:
                return self.i+1
        return -1
a=linear_Search()
n=a.search_num(5,[4,6,7,9,8,0,6,5])
print(f"{n}")
```
