```python
import tkinter as tk

class gui:
    def __init__(self):
        self.w = tk.Tk()
        self.w.geometry("500x500")
        self.w.title("GUI Python")
        
        self.label = tk.Label(self.w, text="Hello Burhan", font=('Arial', 20))
        self.label.pack(padx=10, pady=10)
        
        self.textbox = tk.Text(self.w, height=1, font=('Arial', 20))
        self.textbox.pack(padx=10, pady=10)
        
        self.gridlayout = tk.Frame(self.w)
        self.gridlayout.columnconfigure(0, weight=1)
        self.gridlayout.columnconfigure(1, weight=1)
        self.gridlayout.columnconfigure(2, weight=1)
        self.gridlayout.columnconfigure(3, weight=1)
        
        self.create_buttons()
        
        self.gridlayout.pack(fill='x')
        
        self.w.mainloop()

    def create_buttons(self):
        buttons = [
            ('1', 0, 0), ('2', 0, 1), ('3', 0, 2), ('4', 0, 3),
            ('5', 1, 0), ('6', 1, 1), ('7', 1, 2), ('9', 1, 3),
            ('0', 2, 0), 
            ('+', 3, 0), ('-', 3, 1), ('x', 3, 2), ('/', 3, 3),
            ('=', 4, 0),('C', 4, 1)
        ]

        for (text, row, col) in buttons:
            button = tk.Button(self.gridlayout, text=text, font=('Arial', 20), command=lambda t=text: self.printmath(t))
            button.grid(row=row, column=col, sticky=tk.W+tk.E)
        
    def printmath(self, i=""):
        if i == "=":
            try:
                # Replace x with * for multiplication and evaluate the expression
                expression = self.textbox.get("1.0", tk.END).strip().replace('x', '*')
                result = eval(expression)
                self.textbox.delete("1.0", tk.END)
                self.textbox.insert(tk.END, str(result))
            except Exception as e:
                self.textbox.delete("1.0", tk.END)
                self.textbox.insert(tk.END, "Error")
        elif i == "C":
            self.textbox.delete("1.0", tk.END)
        else:
            self.textbox.insert(tk.END, i)

g = gui()
```
