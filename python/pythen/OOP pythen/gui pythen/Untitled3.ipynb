```python
import mysql.connector
from mysql.connector import Error

class Outer:
    def make_connection(self):
        try:
            # Establish a connection to the MySQL database
            connection = mysql.connector.connect(
                host='localhost',
                database='MedMS',
                user='Burhan',
                password='1234'
            )

            if connection.is_connected():
                cursor = connection.cursor()
                # Execute the SQL query
                cursor.execute("SELECT * FROM user")
                rows = cursor.fetchall()

                # Loop through the rows and print each column
                for row in rows:
                    col1_value = row[0]  # Assuming the first column is Id
                    col2_value = row[1]  # Assuming the second column is Name
                    col3_value = row[3]  # Assuming the fourth column is Email (Note: Index is 0-based)
                    print(f"Id: {col1_value} Name: {col2_value} Email: {col3_value}")

        except Error as e:
            print("Error while connecting to MySQL", e)

        finally:
            # Ensure the cursor and connection are closed
            if connection.is_connected():
                cursor.close()
                connection.close()
                print("MySQL connection is closed")

if __name__ == "__main__":
    outer = Outer()
    outer.make_connection()
```
