```python
import tkinter as tk
from tkinter import messagebox

class Calculator:
    def __init__(self):
        self.window = tk.Tk()
        self.window.geometry("500x380")
        self.window.title("Python GUI Calculator")

        self.current_input = ""
        self.total = 0
        self.current_operator = None

        self.create_widgets()
        self.window.mainloop()

    def create_widgets(self):
        self.label = tk.Label(self.window, text="Calculator", font=('Arial', 20))
        self.label.pack(padx=20, pady=20)

        self.textbox = tk.Entry(self.window, font=('Arial', 20), justify='right')
        self.textbox.pack(padx=10, pady=10, fill=tk.BOTH)

        self.buttonframe = tk.Frame(self.window)
        self.buttonframe.pack(fill='x')

        self.create_buttons()

    def create_buttons(self):
        buttons = [
            ('7', self.on_number), ('8', self.on_number), ('9', self.on_number), ('/', self.on_operator),
            ('4', self.on_number), ('5', self.on_number), ('6', self.on_number), ('*', self.on_operator),
            ('1', self.on_number), ('2', self.on_number), ('3', self.on_number), ('-', self.on_operator),
            ('C', self.on_clear), ('0', self.on_number), ('=', self.on_equal), ('+', self.on_operator),
        ]

        for i, (text, command) in enumerate(buttons):
            row, col = divmod(i, 4)
            button = tk.Button(self.buttonframe, text=text, font=('Arial', 20), command=lambda t=text: command(t))
            button.grid(row=row, column=col, sticky=tk.W+tk.E)

    def on_number(self, num):
        self.current_input += num
        self.update_textbox()

    def on_operator(self, operator):
        if self.current_input:
            self.total = float(self.current_input)
        self.current_operator = operator
        self.current_input = ""
        self.update_textbox()

    def on_clear(self):
        self.current_input = ""
        self.total = 0
        self.current_operator = None
        self.update_textbox()

    def on_equal(self):
        if self.current_operator and self.current_input:
            if self.current_operator == '+':
                self.total += float(self.current_input)
            elif self.current_operator == '-':
                self.total -= float(self.current_input)
            elif self.current_operator == '*':
                self.total *= float(self.current_input)
            elif self.current_operator == '/':
                if float(self.current_input) != 0:
                    self.total /= float(self.current_input)
                else:
                    messagebox.showerror("Error", "Division by zero")
                    self.on_clear()
                    return
            self.current_input = str(self.total)
            self.current_operator = None
            self.update_textbox()

    def update_textbox(self):
        self.textbox.delete(0, tk.END)
        self.textbox.insert(tk.END, self.current_input)

cal = Calculator()
```
