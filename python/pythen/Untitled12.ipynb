```python
class OuterClass:
    def __init__(self,x):
        self.x=x
    def outer_method(self):
        print("this is the outer class")
        
    class InnerClass:
        def __init__(self,y):
            self.y=y
        def inner_method(self):
            print("this is the inner method")
            
outer_obj=OuterClass(5)
outer_obj.outer_method()

inner_obj=outer_obj.InnerClass(10)
inner_obj.inner_method()

print("outer class attribute",outer_obj.x)

print("outer class attribute",inner_obj.y)


        
        
```
