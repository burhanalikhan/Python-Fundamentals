```python
class student:
    def __init__(self,name="",enrol=0,GPA=0.0):
        self.name=name
        self.enrol=enrol
        self.GPA=GPA
s1=student("Burhan",1809,3.5)
print(f"Name : {s1.name} , Enrolment Number : {s1.enrol} , GPA : {s1.GPA}")
```

```python
class employee:
    def __init__(self,name="",enrol=0,Designation=""):
        self.name=name
        self.enrol=enrol
        self.Designation=Designation
e1=employee("Burhan",1809,"S")
print(f"Name : {e1.name} , Enrolment Number : {e1.enrol} , Designation : {e1.Designation}")
```

```python
class SoloLeveling:
    def __init__(self,name="",enrol=0,Rank="",Guild=""):
        self.name=name
        self.enrol=enrol
        self.Rank=Rank
        self.Guild=Guild
hunter1=SoloLeveling()
hunter1.name="Burhan Ali Khan"
hunter1.enrol="01"
hunter1.Rank="S"
hunter1.Guild="Ahgin"
print(f"Hunter Name : {hunter1.name} , Number : {hunter1.enrol} , Rank : {hunter1.Rank} , Guild = {hunter1.Guild}")
```
