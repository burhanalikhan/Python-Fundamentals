```python
#input name
name=input("what is your name ").strip().title()

#split the str in order you want
name_parts = name.split()

# Check the number of parts
if len(name_parts) == 3:
    first, middle, last = name_parts
elif len(name_parts) == 2:
    first, last = name_parts
    middle = ""
elif len(name_parts) == 1:
    first = name_parts[0]
    middle = ""
    last = ""
else:
    print("Error: Name format not recognized")
    first = ""
    middle = ""
    last = ""

# Printing name
print(f"Hello, {first}")
```

```python
def main():
    number=get_number()
    meow(number)
    
def get_number():
    while True:
        i=int(input("What n?"))
        if i > 0:
            break
    return i
    
def meow(n):
    for _ in range(n):
        print("meow")
        
main()
```
