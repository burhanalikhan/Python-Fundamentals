```python
try:
    x=int(input("What is x "))
except ValueError:
    print("x is not an integer")
else:
    print(f"x is {x}")
```

```python
while True:
    try:
        x=int(input("What is x "))
    except ValueError:
        print("x is not an integer")
        continue
    else:
        print(f"x is {x}")
        break
```

```python
def main():
    x=get_int("What is your age? ")
    print(f"age is {x}")
def get_int(p):
    while True:
        try:
            return int(input(p))
        except ValueError:
            pass
    
main()
```

```python
def main():
    c=user_defined_Exception("Low Bank balance",800)
    try :
        c.Withdraw(900)
    except user_defined_Exception as e:
        print(f"An error occured {e}")
class user_defined_Exception(Exception):
    def __init__(self,message,money=0):
        self.money=money
        self.message=message
        super().__init__(self.message)
    def Withdraw(self,num):
        if num > self.money:
            raise user_defined_Exception("Low Bank balance")
        else:
            print("Withdraw Succesfully")
            
main()
```
