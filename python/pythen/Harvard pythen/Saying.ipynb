```python
# Creating a Python module file
module_code = """
def main():
    hello("world")
    goodbye("world")
def hello(name):
    print(f"hello {name}")
def goodbye(name):
    print(f"goodeyr {name}")
"""

with open('saying.py', 'w') as file:
    file.write(module_code)

print("Module 'saying.py' created successfully.")
```
