```python
from abc import ABC,abstarctmethod

class Animal(ABC):
    def eat(self):
        pass
    def move(self):
        pass


class Dog(Animal):
    def eat(self):
        print("dog is eating")
    def move(self):
        print("dog is moving")
dog=Dog()
dog.eat()
dog.move()
```
